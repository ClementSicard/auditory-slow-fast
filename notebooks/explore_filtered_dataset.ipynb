```python
import pandas as pd

path = "../data/epic-kitchens-100-annotations/filtered_train.pkl"
```

```python
df = pd.read_pickle(path)
df.head()
```

```python
df[["precs_vec", "precs"]]

precs = set()
posts = set()
for i in df["precs"]:
    precs.update([a.replace("not-", "") for a in i])
for i in df["posts"]:
    posts.update([a.replace("not-", "") for a in i])

sorted(precs.union(posts))
```
