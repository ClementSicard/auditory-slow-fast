```python
import os
from loguru import logger
import librosa
import soundfile as sf
from IPython.display import Audio, display
```

```python
def trim(input_path: str, output_path: str, start_time: float, end_time: float):
    """
    Trim an audio file to a specified time interval.
    """
    y, sr = librosa.load(input_path, sr=None)  # 'None' means no resampling
    y = y[int(start_time * sr) : int(end_time * sr)]
    sf.write(output_path, y, sr)
    logger.info(f"Trimmed {input_path} to {output_path}.")
```

```python
file_path = "../data/EPIC-KITCHENS/P10/videos/P10_04.wav"

if not os.path.exists(file_path):
    logger.error("File not found: {}".format(file_path))
else:
    logger.success("File found: {}".format(file_path))
```

```python
from datetime import timedelta

output_path = "../data/EPIC-KITCHENS/P10/videos/P10_04_trimmed.wav"
start_ts = timedelta(
    hours=0,
    minutes=10,
    seconds=3,
    milliseconds=0
).total_seconds()
end_ts = timedelta(
    hours=0,
    minutes=10,
    seconds=5,
    milliseconds=0
).total_seconds()

trim(
    file_path,
    output_path=output_path,
    start_time=start_ts,
    end_time=end_ts,
)

display(Audio(output_path))
```
